```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from sqlalchemy import create_engine, text
from urllib.parse import quote_plus
print("‚úÖsucesso.")
```

```python
df = pd.read_csv('FPI.csv', sep=',') # extraindo o arquivo csv e definindo como delimitador a ,
print("‚úÖsucesso.")
```

```python
df = df.iloc[:, :12]  # Mant√©m apenas as 12 primeiras colunas
print(f"‚úÖ Removidas as colunas 13 e 14. Total atual: {len(df.columns)} colunas.")
```

```python
df.info() # Para ver  o tipo de cada coluna e se h√° valores nulos
print("‚úÖsucesso.")
```

```python
df = df.rename(columns={
    'Carimbo de data/hora': 'data_hora',
    'Nome de usu√°rio': 'usuario',
    'Qual √© a sua fun√ß√£o na √°rea da constru√ß√£o civil?\n(ex: engenheiro, mestre de obras, arquiteto, cliente, outro)': 'funcao_area',
    'Com que frequ√™ncia voc√™ acompanha o andamento de obras?': 'frequencia_de_acompanhamento',
    'Atualmente, como voc√™ registra ou acompanha o progresso das obras?': 'registro_acompanhamento',
    'Quais informa√ß√µes voc√™ considera mais importantes em uma obra?': 'info_importantes',
    '  Voc√™ costuma ter dificuldades em organizar ou acessar informa√ß√µes sobre o andamento das obras?  ': 'dificuldade_em_organizar_acessar',
    '  Voc√™ considera √∫til um aplicativo que registre fotos e dados de cada etapa da obra em um s√≥ lugar?  ': 'considera_app_util',
    'Quais recursos ou funcionalidades voc√™ gostaria que esse aplicativo tivesse?  ': 'recursos_funcionalidades',
    'Como voc√™ compartilha atualiza√ß√µes do andamento da obra com equipe e clientes?': 'atualizacoes_e_andamento',
    '  Voc√™ utiliza algum aplicativo ou software para gest√£o de obras? Se sim, qual?  ': 'usa_app_software',
    '  Quais dificuldades voc√™ enfrenta no acompanhamento ou comunica√ß√£o durante o andamento da obra?  ': 'dificuldades_acompanhamento'
})

print("‚úÖsucesso.")
```

```python
print(df.columns.tolist())
print("‚úÖsucesso.")
```

```python
df["data_hora"] = pd.to_datetime(df["data_hora"], dayfirst=True, errors="coerce") # define que a coluna √© de data
df['data_hora'] = df['data_hora'].astype('datetime64[ns]')
df["data_hora"] = df["data_hora"].dt.strftime("%d/%m/%Y  %H:%M")  # muda a formata√ß√£o da datada data para o formato do brasi
print("‚úÖsucesso.")
```

```python
df.fillna({
    'dificuldades_acompanhamento': 'N√£o informado',
}, inplace=True)
print("‚úÖsucesso.")
```

```python
# Padroniza√ß√£o da coluna 'profiss√£o'
df['funcao_area'] = (
    df['funcao_area']
    .str.strip()  # remove espa√ßos extras
    .replace({
        # Engenheiros(as)
        'Engenheiro civil': 'Engenheiro(a)',
        'Engenheiro Civil': 'Engenheiro(a)',
        'Engenheiro ': 'Engenheiro(a)',
        'Engenheiro': 'Engenheiro(a)',
        'engenheiro': 'Engenheiro(a)',
        'engenheiro ': 'Engenheiro(a)',
        'Engenheira': 'Engenheiro(a)',
        'Engenheira ': 'Engenheiro(a)',
        'Engenheira civil': 'Engenheiro(a)',
        'Engenheira Civil': 'Engenheiro(a)',

        # Arquitetos(as)
        'Arquiteto': 'Arquiteto(a)',
        'Arquiteto ': 'Arquiteto(a)',
        'Arquiteta': 'Arquiteto(a)',
        'Arquiteta ': 'Arquiteto(a)',

        # Mestre de obras
        'Mestre de obras': 'Mestre de obras',
        'Mestre de obras (ou equivalente).': 'Mestre de obras',
        'Mestre de obra': 'Mestre de obras',
        'Mestre de obras ': 'Mestre de obras',
        'Mestre de obra ': 'Mestre de obras',
        'Gestor de Obras': 'Mestre de obras',

        # Clientes
        'Cliente': 'Cliente',
        'Cliente ': 'Cliente',

        # Coordenadores / gestores
        'Coordenador Regional de Vendas T√©cnicas ': 'Coordenador',
        'Coordenador de Data Analytics': 'Coordenador',
        'Coordenar': 'Coordenador(a)',

        # Outros
        'Pedreiro': 'Pedreiro',
        'Astucioso ': 'Astucioso',
        'Estudante': 'Estudante',
        'Outro ': 'Outro',
        'Outros ': 'Outro',
        'Outros': 'Outro',
        'Consultor de SGQ': 'Consultor',
    })
)

print("‚úÖ Coluna 'profiss√£o' padronizada com sucesso!")

```

```python
df.to_csv('FPI_limpo.csv', index=False)  # Salvando o DataFrame limpo em um novo arquivo CSV
print("‚úÖ Arquivo CSV limpo salvo com sucesso como 'FPI_limpo.csv'")
```
